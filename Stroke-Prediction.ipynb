```python
x= data.drop("stroke", axis=1)
x.head()
```

```python
y=data[["stroke"]]
y.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
x= scaler.fit_transform(x)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.33, random_state= 42)
```

```python
x_train
```

Model Training (DT)

```python
from sklearn.tree import DecisionTreeClassifier
dtm= DecisionTreeClassifier(criterion = 'entropy', max_depth = 15)
dtm.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
X_train_prediction_dtm = dtm.predict(x_train)
accuracy_score( y_train,X_train_prediction_dtm,)
print('Accuracy on training data : ', round(accuracy_score(X_train_prediction_dtm, y_train,)*100,2), '%')
```

```python
#accuracy
dtm.score(x_test,y_test)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier

# Define max depths to try
max_depths = np.arange(1, 21)
train_accuracies = []
test_accuracies = []

# Train and test decision tree with different max depths
for depth in max_depths:
    decision = DecisionTreeClassifier(max_depth=depth, random_state=1)
    decision.fit(x_train, y_train)
    train_accuracy = decision.score(x_train, y_train)
    test_accuracy = decision.score(x_test, y_test)
    train_accuracies.append(train_accuracy)
    test_accuracies.append(test_accuracy)

# Plot the accuracy graph
plt.plot(max_depths, train_accuracies, label='Training Accuracy')
plt.plot(max_depths, test_accuracies, label='Testing Accuracy')
plt.xlabel('Max Depth')
plt.ylabel('Accuracy')
plt.title('Decision Tree Accuracy')
plt.legend()
plt.show()

```
